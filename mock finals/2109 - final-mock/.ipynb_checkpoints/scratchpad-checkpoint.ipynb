### 2. Detection and Handling of Missing Values

```python
pd.options.display.max_rows = 500

# ----------------TABULATION---------------
# remove V38 and V39
columns_to_remove = ['V38', 'V39', 'target']
new_df = df.drop(columns = columns_to_remove)
y = df['target']

# interpolate V5
nan_mask = new_df['V5'].isna()
non_nan_values = new_df['V5'][~nan_mask]
mean, std = np.mean(non_nan_values), np.std(non_nan_values)
random_values = np.random.normal(loc=mean, scale=std, size=nan_mask.sum())
new_df = new_df.copy()
new_df.loc[nan_mask, 'V5'] = random_values


# interpolate V15
nan_mask = new_df['V15'].isna()
non_nan_values = new_df['V15'][~nan_mask]
mean, std = np.mean(non_nan_values), np.std(non_nan_values)
random_values = np.random.normal(loc=mean, scale=std, size=nan_mask.sum())
new_df = new_df.copy()
new_df.loc[nan_mask, 'V15'] = random_values

# Drop any rows in the df that have NAN in any column that has a NAN
new_df = new_df.dropna(how="any")
# new_df

# # Finding rows that have missing values
# nan_count_per_row = new_df.isnull().sum()
# nan_count_per_row


# ----------------IMAGES---------------------

# Remove columns 19 and 24
columns_to_remove = ['19', '24', '61', '62', '63']
images = images.drop(columns = columns_to_remove)

# Interpolate NaN values in the 0th column
nan_mask = images['0'].isna()
non_nan_values = images['0'][~nan_mask]
mean, std = np.mean(non_nan_values), np.std(non_nan_values)
random_values = np.random.normal(loc=mean, scale=std, size=nan_mask.sum())
images = images.copy()
images.loc[nan_mask, '0'] = random_values

# Interpolate NaN values in the 12th column
nan_mask = images['12'].isna()
non_nan_values = images['0'][~nan_mask]
mean, std = np.mean(non_nan_values), np.std(non_nan_values)
random_values = np.random.normal(loc=mean, scale=std, size=nan_mask.sum())
images = images.copy()
images.loc[nan_mask, '0'] = random_values

# Drop all columns with NAN inside
images = images.dropna(how="any")

new_df
```

### 3. Detection and Handling of Outliers

```python
from scipy import stats

# Replace outliers (Z-Score = 3 standard deviations from the mean) with NAN
for column in new_df.columns:
    if np.issubdtype(new_df[column].dtype, np.number):  # Check if the column is numeric
        z_scores = np.abs(stats.zscore(new_df[column]))
        threshold = 3  # You can adjust the threshold as needed
        new_df[column] = np.where(z_scores > threshold, np.nan, new_df[column])

# Drop any rows in the df that have NAN in any column that has a NAN
new_df = new_df.dropna(how="any")
# new_df.describe()


# # Finding rows that have missing values
# nan_count_per_row = images.isnull().sum()
# nan_count_per_row

# ----------- IMAGES ----------------
# Replace outliers (Z-Score = 3 standard deviations from the mean) with NAN
for column in images.columns:
    if np.issubdtype(images[column].dtype, np.number):  # Check if the column is numeric
        z_scores = np.abs(stats.zscore(images[column]))
        threshold = 3  # You can adjust the threshold as needed
        images[column] = np.where(z_scores > threshold, np.nan, images[column])

# Drop any rows in the df that have NAN in any column that has a NAN
images = images.dropna(how="any")
# images

new_df
```

```python
# Old Code
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler

# Define the neural network model
class NeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, output_size, dropout_rate):
        super(NeuralNetwork, self).__init__()
        self.layer1 = nn.Linear(input_size, hidden_size1)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(p=dropout_rate)
        self.layer2 = nn.Linear(hidden_size1, hidden_size2)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(p=dropout_rate)
        self.layer3 = nn.Linear(hidden_size2, output_size)

    def forward(self, x):
        x = self.layer1(x)
        x = self.relu1(x)
        x = self.dropout1(x)
        x = self.layer2(x)
        x = self.relu2(x)
        x = self.dropout2(x)
        x = self.layer3(x)
        return x

# Separating the Data into Training and Validation
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
X_val_tensor = torch.tensor(X_test.values, dtype=torch.float32)
y_val_tensor = torch.tensor(y_test.values, dtype=torch.float32)

# Example: Assuming input size is 10, hidden size is 64, and output size is 1
input_size = X_train_tensor.shape[1]
hidden_size1 = 128
hidden_size2 = 64
output_size = 1
dropout_rate = 0.2

# Instantiate the model
model = NeuralNetwork(input_size, hidden_size1, hidden_size2, output_size, dropout_rate)

# Loss Function and Optimisers
criterion = nn.MSELoss()
# Define L2 regularization (weight decay)
weight_decay = 1e-4  # You can adjust this value
optimizer = optim.Adam(model.parameters(), lr=0.01, weight_decay=weight_decay)


# Training the Model
num_epochs = 100
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X_train_tensor)
    
    # Reshape your_output_data to match the shape of the predicted output
    y_train_tensor = y_train_tensor.view(-1, 1)
    
    loss = criterion(outputs, y_train_tensor)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Print the loss for every epoch (you can customize this based on your needs)
    print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item()}')
```
